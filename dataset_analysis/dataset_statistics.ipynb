```python
import pandas as pd

from Bio import SeqIO

def parse_fasta_biopython(filename):
    records = []
    # Use 'with open' to ensure the file is closed properly
    with open(filename, "r") as handle:
        for seq_record in SeqIO.parse(handle, "fasta"):
            if len(seq_record.seq) > 2000 and len(seq_record.seq) < 20000:
                records.append(seq_record)
            # You can access attributes like:
            # print(seq_record.id)
            # print(seq_record.seq)
            # print(len(seq_record))
    return records

fasta_file = parse_fasta_biopython("/home/Users/fmq1/synthetic_sequences.fasta")
SeqIO.write(fasta_file,"/home/Users/fmq1/synthetic_sequences_filtered.fasta","fasta")
len(fasta_file)
```
